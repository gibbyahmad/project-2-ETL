```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
json_file = "./AllCards.json"
mtg_df = pd.read_json(json_file)
mtg_df
```

```python
#transpose dataframe (so card names are in the index columnar, and then make that index into a column)
mtg_df = pd.DataFrame(mtg_df.transpose())
mtg_df.reset_index(level=0, inplace=True)
mtg_df
```

```python
##Clean Dataframe, rename card name column 'cardname'
clean_mtg_df = mtg_df[["index","convertedManaCost", "text", "type"]]
clean_mtg_df = clean_mtg_df.rename(columns={"index":"cardname"})
clean_mtg_df.head()
```

##creating SQLite file and pushing data into it

```python
# Imports the method used for connecting to DBs
from sqlalchemy import create_engine

# Imports the methods needed to abstract classes into tables
from sqlalchemy.ext.declarative import declarative_base

# Allow us to declare column types
from sqlalchemy import Column, Integer, String, Float 
```

```python
# Sets an object to utilize the default declarative base in SQL Alchemy
Base = declarative_base()
```

```python
# Creates a connection to our DB
engine = create_engine("sqlite:///mtgdata.sqlite")
conn = engine.connect()
```

```python
# Create (if not already in existence) the tables associated with our classes.
Base.metadata.create_all(engine)
```

```python
# Session is a temporary binding to our D()
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
Base.metadata.drop_all(engine)
clean_mtg_df.to_sql('cardinfo', con=engine)
engine.execute("SELECT * FROM cardinfo").fetchall()
```
